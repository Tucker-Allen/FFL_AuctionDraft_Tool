### Fantasy Pros

```python
with open('./fantasypros_values.txt') as f:
    lines = f.readlines()
```

```python
lines[0].split('\t')
```

```python
fpros_players = {}
for line in lines:
    sep = line.split('\t')
    if sep[1] in ['QB', 'RB', 'WR', 'DST', 'TE', 'K']:
        pos = sep[1]
        continue
    fpros_players[ sep[1] ] = {'name': sep[1],
                              'position': pos,
                              'fpros_pre_draft$': int(sep[2].replace('$', '').strip())}
```

```python
fpros_players
```

```python
len(fpros_players)
```

### ESPN Values

```python
with open('./espn_values.txt') as f:
    lines2 = f.readlines()
```

```python
lines2[4].split(') ')
```

```python
espn_players = {}
for line in lines2:
    for one_entry in line.split(') ')[1:]:
        try:
            final_split = one_entry.split(' $')
            player = final_split[0]
            pre_draft = int(final_split[1][:2].strip())

            espn_players[player] = {'name': player,
                                   'espn_pre_draft$': pre_draft}
        except:
            print('!!!!!!!!!!!!!!', one_entry)
        
```

```python
# {espn_name: fpros_name}
```

```python
espn_to_fpros = {'Todd Gurley II, LAR': 'Todd Gurley, LAR',
                'Mark Ingram II, BAL': 'Mark Ingram, BAL',
                'DJ Moore, CAR': 'D.J. Moore, CAR',
                'Marvin Jones Jr., DET': 'Marvin Jones, DET',
                'Mitchell Trubisky, CHI': 'Mitch Trubisky, CHI',
                'Duke Johnson Jr., HOU': 'Duke Johnson, HOU',
                'Ronald Jones, TB': 'Ronald Jones II, TB',
                'Will Fuller V, HOU': 'Will Fuller, HOU',
                'Ted Ginn Jr., NO': 'Ted Ginn, NO'}
```

```python
for player in espn_players:
    if player in espn_to_fpros:
        sub_player = espn_to_fpros[player]
        fpros_players[sub_player]['espn_pre_draft$'] = espn_players[player]['espn_pre_draft$']
        continue
    try:
        fpros_players[player]['espn_pre_draft$'] = espn_players[player]['espn_pre_draft$']
    except:
        print('Skipping ', player)
```

```python
all_players = fpros_players.copy()
```

```python
for player in all_players:
    try:
        all_players[player]['espn_pre_draft$'] = all_players[player]['espn_pre_draft$']
        all_players[player]['fpros_pre_draft$'] = all_players[player]['fpros_pre_draft$']
    except:
        print('Setting ', player, 'to None')
        all_players[player]['espn_pre_draft$'] = None
        all_players[player]['fpros_pre_draft$'] = all_players[player]['fpros_pre_draft$']
```

```python
for player in all_players:
    all_players[player]['tier'] = None
```

```python
for player in all_players:
    if not all_players[player]['espn_pre_draft$']:
        continue
    
    if all_players[player]['position'] == 'RB':
        if all_players[player]['espn_pre_draft$'] > 1:
            all_players[player]['tier'] = 'tier_3'
        if all_players[player]['espn_pre_draft$'] > 30:
            all_players[player]['tier'] = 'tier_2'
        if all_players[player]['espn_pre_draft$'] > 50:
            all_players[player]['tier'] = 'tier_1'
            
    if all_players[player]['position'] == 'WR':
        if all_players[player]['espn_pre_draft$'] > 0:
            all_players[player]['tier'] = 'tier_4'
        if all_players[player]['espn_pre_draft$'] > 9:
            all_players[player]['tier'] = 'tier_3'
        if all_players[player]['espn_pre_draft$'] > 29:
            all_players[player]['tier'] = 'tier_2'
        if all_players[player]['espn_pre_draft$'] > 40:
            all_players[player]['tier'] = 'tier_1'
            
    if all_players[player]['position'] == 'TE':
        if all_players[player]['espn_pre_draft$'] > 0:
            all_players[player]['tier'] = 'tier_3'
        if all_players[player]['espn_pre_draft$'] > 5:
            all_players[player]['tier'] = 'tier_2'
        if all_players[player]['espn_pre_draft$'] > 25:
            all_players[player]['tier'] = 'tier_1'
```
